<h3> Question 3 </h3>

Montrez que le problème primal résolu par le SVM peut se réécrire de la façon suivante : 

$$\underset{\textbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}}{\operatorname{argmin}} \left( \frac{1}{2} || \textbf{w} ||^{2} + C \sum_{i = 1}^{n} \left[ 1 - y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right) \right]_{+} \right)$$

<hr/>

Notons $(1)$ le problème primal résolu par le SVM (introduit dans l'énoncé) et $(2)$ le problème de la question.

Afin de simplifier les notations, introduisons 

$$C \left( w_{0} \right) = \left\{ \xi \in \mathbb{R}^{n} : 
\left\{ \begin{array}
\forall \forall i \in [\![ 1, n ]\!], & \xi_{i} \geq 0 \\
        \forall i \in [\![ 1, n ]\!], & y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right) \geq 1 - \xi_{i} \end{array} \right. \right\} $$

et $$A = \min_{\textbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}, \xi \in C(w_{0})} \left( \frac{1}{2} || \textbf{w} ||^{2} + C \sum_{i = 1}^{n} \xi_{i} \right)$$ 

On a $$A = \min_{\textbf{w} \in \mathcal{H}} \frac{1}{2} || \textbf{w} ||^{2} + C \min_{\textbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}, \xi \in C(w_{0})} \sum_{i = 1}^{n} \xi_{i} $$ car $C \geq 0$.

De plus, $$A = \min_{\textbf{w} \in \mathcal{H}} \frac{1}{2} || \textbf{w} ||^{2} + C \min_{\textbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}} \sum_{i = 1}^{n} \left[ 1 - y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right) \right]_{+}$$ par définition de l'ensemble $C(a)$.

Ainsi, les variables étant toujours séparables, on a $$A = \min_{\textbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}} \left( \frac{1}{2} || \textbf{w} ||^{2} + C \sum_{i = 1}^{n} \left[ 1 - y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right) \right]_{+} \right)$$

Et les minima sont donc égaux, i.e. : $$\boxed{\min_{\textbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}, \xi \in C(w_{0})} \left( \frac{1}{2} || \textbf{w} ||^{2} + C \sum_{i = 1}^{n} \xi_{i} \right) = \min_{\textbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}} \left( \frac{1}{2} || \textbf{w} ||^{2} + C \sum_{i = 1}^{n} \left[ 1 - y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right) \right]_{+} \right)}$$

Soit $\left( \textbf{w}^{*}_{1}, w_{0, 1}^{*}, \xi^{*}_{1} \right)$ un minimiseur du problème $(1)$. On remarque que le couple $(\textbf{w}^{*}_{2}, w_{0, 2}^{*}) = \left( \textbf{w}^{*}_{1}, w_{0, 1}^{*} \right)$ est un minimiseur du problème $(2)$.

Réciproquement, soit $\left( \textbf{w}^{*}_{2}, w_{0, 2}^{*} \right)$ un minimiseur du problème $(2)$. En posant, pour $i \in [\![ 1, n ]\!], \xi_{i} = \left[ 1 - y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right) \right]_{+}$, on a bien $\xi \in C \left( w_{0, 2}^{*} \right)$ et on trouve que le triplet $(\textbf{w}^{*}_{1}, w_{0, 1}^{*}, \xi^{*}_{1}) = (\textbf{w}^{*}_{2}, w_{0, 2}^{*}, \xi)$ est un minimiseur du problème $(1)$. Ainsi,

<p style="text-align:center";> <span style="border:1px solid black;padding:1%">On peut construire un optimum de $(1)$ à partir d'un optimum de $(2)$ et vice versa.</span> </p>

On peut donc conclure :

<p style="text-align:center";> <span style="border:1px solid black;padding:1%">Le problème primal peut se réécrire de la façon énoncée ci-dessus.</span> </p>

<h3> Question 4 </h3>

Expliquez la phrase : "un SVM minimise l'erreur de classification à l'aide d'un majorant convexe de la fonction qui vaut $1$ quand la marge est négative et $0$ sinon". La fonction $x \mapsto \left[ 1 - x \right]_{+} = max \left( 0, 1 - x \right)$ est appelée <i>Hinge</i> (charnière en français).

<hr/>

La marge correspond, pour $i \in [\![ 1, n ]\!]$, à la quantité $y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right)$. Si cette quantité est positive, le point $\left( x_{i}, y_{i} \right)$ est bien classé. Si elle est négative, le point est mal classé. Ainsi, la fonction "qui vaut $1$ quand la marge est négative et $0$ sinon" correspond à l'erreur 0-1 et est donnée par $\mathbb{1}_{y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right) \leq 0}$.

Par ailleurs, on remarque que si $x \leq 0$, on a 

$$\mathbb{1}_{x \leq 0} = 1 \leq 1 - x = \left[ 1 - x \right]_{+}$$

Et si $x > 0$, 

$$\mathbb{1}_{x \leq 0} = 0 \leq \left[ 1 - x \right]_{+}$$

Ainsi, la fonction Hinge est un majorant de l'erreur 0-1. Comme cette fonction est convexe (son épigraphe est convexe), il s'agit donc d'un majorant convexe de l'erreur 0-1. Or le SVM minimise le terme $$\sum_{i = 1}^{n} \left[ 1 - y_{i} \left( \langle \textbf{w}, \phi(x_{i}) \rangle + w_{0} \right) \right]_{+}$$ qui correspond en fait à une somme de fonctions Hinge évaluées en la marge de chacun des points. On peut donc en conclure : 

<p style="text-align:center";> <span style="border:1px solid black;padding:1%"> un SVM minimise l'erreur de classification à l'aide d'un majorant convexe de la fonction qui vaut $1$ quand la marge est négative et $0$ sinon </span> </p>

```python
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4,
                              color=True, funneled=False, slice_=None,
                              download_if_missing=True)


# introspect the images arrays to find the shapes (for plotting)
images = lfw_people.images / 255.
n_samples, h, w, n_colors = images.shape

# the label to predict is the id of the person
target_names = lfw_people.target_names.tolist()


# Pick a pair to classify such as
names = ['Tony Blair', 'Colin Powell']
# names = ['Donald Rumsfeld', 'Colin Powell']

idx0 = (lfw_people.target == target_names.index(names[0]))
idx1 = (lfw_people.target == target_names.index(names[1]))

images = np.r_[images[idx0], images[idx1]]
n_samples = images.shape[0]
y = np.r_[np.zeros(np.sum(idx0)), np.ones(np.sum(idx1))].astype(np.int)


# Extract features

# features using only illuminations
X = (np.mean(images, axis=3)).reshape(n_samples, -1)

# or compute features using colors (3 times more features)
# X = images.copy().reshape(n_samples, -1)

# Scale features
X -= np.mean(X, axis=0)
X /= np.std(X, axis=0)


# Split data into a half training and half test set
# X_train, X_test, y_train, y_test, images_train, images_test = \
#    train_test_split(X, y, images, test_size=0.5, random_state=0)
# X_train, X_test, y_train, y_test = \
#    train_test_split(X, y, test_size=0.5, random_state=0)

indices = np.random.permutation(X.shape[0])
train_idx, test_idx = indices[:int(X.shape[0]/2)], indices[int(X.shape[0]/2):]
X_train, X_test = X[train_idx, :], X[test_idx, :]
y_train, y_test = y[train_idx], y[test_idx]
images_train, images_test = images[train_idx, :, :, :], images[test_idx, :, :, :]
```

```python
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)


print("Predicting the people names on the testing set")
t0 = time()

# TODO : predict labels for the X_test images
y_pred = clf.predict(X_test)


print("done in %0.3fs" % (time() - t0))
print("Accuracy : %s" % clf.score(X_test, y_test))
```

```python
# Look at the coefficients
pl.figure()
pl.imshow(np.reshape(clf.coef_, (h, w)))
```

```python
# Qualitative evaluation of the predictions using matplotlib

def plot_gallery(images, titles, n_row=3, n_col=4):
    """Helper function to plot a gallery of portraits"""
    pl.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    pl.subplots_adjust(bottom=0, left=.01, right=.99, top=.90,
                       hspace=.35)
    for i in range(n_row * n_col):
        pl.subplot(n_row, n_col, i + 1)
        pl.imshow(images[i])
        pl.title(titles[i], size=12)
        pl.xticks(())
        pl.yticks(())


def title(y_pred, y_test, names):
    pred_name = names[int(y_pred)].rsplit(' ', 1)[-1]
    true_name = names[int(y_test)].rsplit(' ', 1)[-1]
    return 'predicted: %s\ntrue:      %s' % (pred_name, true_name)



prediction_titles = [title(y_pred[i], y_test[i], names)
                     for i in range(y_pred.shape[0])]

plot_gallery(images_test, prediction_titles)
pl.show()
```

<h3> Question 6 </h3>

En ajoutant des variables de nuisances, augmentant ainsi le nombre de variables à nombre de points d'apprentissage fixé, montrez que la performance chute.

<hr/>

```python
nb_noise_features = np.arange(0, 12000, 2000)
scores = []

noise_features = np.random.randn(X.shape[0], nb_noise_features[-1])

for i in nb_noise_features:
    
    noise = noise_features[:, :i]
    
    tmp_mean = noise.mean(axis = 0)
    tmp_std = noise.std(axis = 0)
    for k in range(i):
        noise[:, k] -= tmp_mean[k]
        noise[:, k] /= tmp_std[k]
    
    X_noise = np.hstack((X, noise))
    X_train_noise, X_test_noise = X_noise[train_idx, :], X_noise[test_idx, :]
    
    clf_noise = SVC(kernel = 'linear')
    clf_noise.fit(X_train_noise, y_train)

    y_pred_noise = clf_noise.predict(X_test_noise)

    scores.append(clf_noise.score(X_test_noise, y_test))
```

```python
plt.figure()
plt.xlabel('Number of noise features')
plt.ylabel('Score of the classifier')
plt.title('Evolution of the performance as a function of the number of noise features')
plt.plot(nb_noise_features, scores)
plt.show()
```

<h3> Question 9 </h3>

En vous basant sur cet exemple écrivez un code qui va calculer la valeur des fonctionnelles primales et duales et vérifiez que les valeurs sont proches. Vous afficherez sur un même graphe les valeurs des fonctionnelles.

<hr/>

La question étant un peu étrange, on propose ici deux méthodes différentes et qui ne font pas la même chose.

<h4> Méthode 1 </h4>

```python
def primal_function(w, w0, C, X_train, y_train):    
    primal = (1/2)*np.linalg.norm(w)**2

    for k in range(len(y_train)):
        primal += C*max(0, 1 - y_train[k]*(w@X_train[k, :] + w0))
    
    return primal[0]
    

def dual_function(alphas, indices, X_train, y_train):
    dual = sum(abs(alphas))
    
    for i in range(len(alphas)):
        for j in range(len(alphas)):
            dual -= (1/2)*abs(alphas[i])*abs(alphas[j])*y_train[indices[i]]*y_train[indices[j]]*(X_train[indices[i], :]@X_train[indices[j], :]) 
    
    return dual
```

```python
C = 0.001
svc_clf = SVC(kernel="linear", C = C)
svc_clf.fit(X_train, y_train)
w = svc_clf.coef_
w0 = svc_clf.intercept_
alphas = svc_clf.dual_coef_[0]
indices = svc_clf.support_

primal_value = primal_function(w, w0, C, X_train, y_train)
dual_value = dual_function(alphas, indices, X_train, y_train)

print('The difference betweeen the primal value and the dual value is: {}'.format(primal_value - dual_value))
```

```python
C_values_bis = [1e-5, 5e-5, 1e-4, 3e-4, 4e-4, 5e-4, 6e-4, 1e-3, 5e-3, 1e-2, 5e-2, 1e-1, 1, 1e2, 1e3]
plot = []

for c in C_values_bis:    
    C = c
    svc_clf = SVC(kernel="linear", C = C)
    svc_clf.fit(X_train, y_train)
    w = svc_clf.coef_
    w0 = svc_clf.intercept_
    alphas = svc_clf.dual_coef_[0]
    indices = svc_clf.support_

    primal_value = primal_function(w, w0, C, X_train, y_train)
    dual_value = dual_function(alphas, indices, X_train, y_train)
    
    plot.append(primal_value - dual_value)
    
plt.figure()
plt.plot(C_values_bis, plot, label = 'Difference between primal and dual value')
plt.xscale('log')
plt.xlabel('C')
plt.ylabel('Difference')
plt.legend(loc = 'best')
plt.show()
```

<h4> Méthode 2 </h4>

```python
plot2 = []

for c in C_values_bis:
    C = c
    lsvc = LinearSVC(C = C, dual = False)
    lsvc.fit(X_train, y_train)

    dlsvc = LinearSVC(C = C, dual = True)
    dlsvc.fit(X_train, y_train)

    w_primal = lsvc.coef_[0]
    w0_primal = lsvc.intercept_

    w_dual = dlsvc.coef_[0]
    w0_dual = dlsvc.intercept_

    plot2.append(primal_function(w_primal, w0_primal, C, X_train, y_train) - primal_function(w_dual, w0_dual, C, X_train, y_train))

plt.figure()
plt.xscale('log')
plt.xlabel('C')
plt.ylabel('Difference')
plt.plot(C_values_bis, plot2, label = 'Difference as a function of C')
plt.show()
```
